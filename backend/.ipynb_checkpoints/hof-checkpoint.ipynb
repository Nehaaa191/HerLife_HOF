```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2, f_classif
from sklearn.impute import SimpleImputer
 
import warnings
warnings.filterwarnings('ignore')
 
print("All libraries imported successfully ✅")
```

```python
df = pd.read_csv('/kaggle/input/datasets/dhruvikarawat/herlife/HerLife_Dataset_1200.csv')
 
print("Shape:", df.shape)
print("\nFirst 3 rows:")
df.head(3)
```

```python
print("=== Column Data Types ===")
print(df.dtypes)
print("\n=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0])
print("\n=== Basic Stats ===")
df.describe()
```

```python
drop_cols = [
    'User_Journal_Entry',
    'Notification_Preference',
    'Most_Energetic_Phase',
    'Menstrual_Product_Used',
]
df.drop(columns=drop_cols, inplace=True)
print("Dropped:", drop_cols)
print("Remaining columns:", df.shape[1])
 
```

```python
preteen_mask  = df['Life_Phase'] == 'Preteen'
menopause_mask = df['Life_Phase'] == 'Menopause'
 
# Columns that are N/A for Preteens
preteen_na_cols = [
    'Flow_Intensity', 'Period_Symptoms', 'PMS_Symptoms',
    'Current_Cycle_Phase', 'Hair_Fall_Level'
]
for col in preteen_na_cols:
    df.loc[preteen_mask, col] = df.loc[preteen_mask, col].fillna('None')
 
# Columns that are N/A for Menopause
menopause_na_cols = [
    'Flow_Intensity', 'Period_Symptoms', 'PMS_Symptoms', 'Hair_Fall_Level'
]
for col in menopause_na_cols:
    df.loc[menopause_mask, col] = df.loc[menopause_mask, col].fillna('None')
 
# PCOS_PCOD_Symptoms — empty means no symptoms
df['PCOS_PCOD_Symptoms'] = df['PCOS_PCOD_Symptoms'].fillna('None')
 
# Skin_Condition_During_Cycle — fill with mode for reproductive, 'Clear' for others
df.loc[~(preteen_mask | menopause_mask), 'Skin_Condition_During_Cycle'] = \
    df.loc[~(preteen_mask | menopause_mask), 'Skin_Condition_During_Cycle'] \
      .fillna(df['Skin_Condition_During_Cycle'].mode()[0])
df['Skin_Condition_During_Cycle'] = df['Skin_Condition_During_Cycle'].fillna('Clear')
 
# Other_Medical_Diagnosis
df['Other_Medical_Diagnosis'] = df['Other_Medical_Diagnosis'].fillna('None')
 
print("Null check after phase-aware fill:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
ordinal_maps = {
    'Exercise_Frequency': {'Never': 0, 'Sometimes': 1, 'Regularly': 2},
 
    'Sleep_Duration': {
        'Less than 5 hrs': 0, '5-7 hrs': 1, '7-9 hrs': 2, 'More than 9 hrs': 3
    },
 
    'Flow_Intensity': {
        'None': 0, 'Light': 1, 'Moderate': 2, 'Heavy': 3, 'Very Heavy': 4
    },
 
    'Overall_Energy_Level': {'Low': 0, 'Medium': 1, 'High': 2},
 
    'Missed_Periods_Frequency': {'Never': 0, 'Occasionally': 1, 'Frequently': 2},
 
    'Caffeine_Intake': {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Daily': 3},
 
    'Hair_Fall_Level': {'None': 0, 'Mild': 1, 'Moderate': 2, 'Severe': 3},
 
    'Skin_Condition_During_Cycle': {
        'Clear': 0, 'Mild breakout': 1, 'Moderate breakout': 2, 'Severe breakout': 3
    },
}
 
for col, mapping in ordinal_maps.items():
    df[col] = df[col].map(mapping)
    unmapped = df[col].isnull().sum()
    if unmapped > 0:
        print(f"  ⚠ {col} has {unmapped} unmapped values → filling with 0")
        df[col] = df[col].fillna(0).astype(int)
    else:
        df[col] = df[col].astype(int)
 
print("Ordinal encoding done ✅")
 
 
```

```python
nominal_cols = [
    'Life_Phase', 'Diet_Type', 'Current_Cycle_Phase', 'Other_Medical_Diagnosis'
]
 
le = LabelEncoder()
label_encoders = {}  # save encoders in case you need to inverse-transform later
 
for col in nominal_cols:
    df[col] = df[col].astype(str)
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
    print(f"  {col}: {dict(zip(le.classes_, le.transform(le.classes_)))}")
 
print("\nNominal encoding done ✅")
```

```python
def expand_symptoms(df, col, prefix):
    """One-hot expand a semicolon-separated symptom column."""
    dummies = df[col].fillna('None').str.get_dummies(sep=';')
    dummies.columns = [f"{prefix}_{c.strip().replace(' ','_')}" for c in dummies.columns]
    if f"{prefix}_None" in dummies.columns:
        dummies.drop(columns=[f"{prefix}_None"], inplace=True)
    return dummies
 
period_dummies = expand_symptoms(df, 'Period_Symptoms',  'SYM')
pms_dummies    = expand_symptoms(df, 'PMS_Symptoms',     'PMS')
pcos_dummies   = expand_symptoms(df, 'PCOS_PCOD_Symptoms', 'PCOS_S')
 
df.drop(columns=['Period_Symptoms', 'PMS_Symptoms', 'PCOS_PCOD_Symptoms'], inplace=True)
df = pd.concat([df, period_dummies, pms_dummies, pcos_dummies], axis=1)
 
print("Symptom columns expanded:")
print("  Period symptoms →", list(period_dummies.columns))
print("  PMS symptoms    →", list(pms_dummies.columns))
print("  PCOS symptoms   →", list(pcos_dummies.columns))
print("\nNew shape:", df.shape)
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
 
# Confirm all numeric
remaining_objects = df.select_dtypes(include='object').columns.tolist()
if remaining_objects:
    print("⚠ Still object dtype:", remaining_objects)
    # Force encode anything left
    for col in remaining_objects:
        df[col] = LabelEncoder().fit_transform(df[col].astype(str))
else:
    print("All columns are numeric ✅")
 
print("Missing values remaining:", df.isnull().sum().sum())
print("Final shape:", df.shape)
```

```python
iqr_cols = ['BMI', 'Weight_kg', 'Water_Intake_Litres', 'Days_Since_Last_Period',
            'Avg_Cycle_Length_days']
 
for col in iqr_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    df[col] = df[col].clip(lower, upper)
    print(f"  {col}: clipped {n_outliers} outliers  [{lower:.1f}, {upper:.1f}]")
 
print("\nOutlier treatment done ✅")
```

```python
df['Cycle_Irregularity_Score'] = (
    (1 - df['Periods_Regular']) +
    df['Missed_Periods_Frequency'] +
    (df['Avg_Cycle_Length_days'].apply(lambda x: 1 if x < 21 or x > 35 else 0))
)
 
# 2. Hormonal stress index (stress × sleep deprivation)
sleep_deprivation = df['Sleep_Duration'].apply(lambda x: 1 if x == 0 else 0)
df['Hormonal_Stress_Index'] = df['Stress_Level_1to5'] * (1 + sleep_deprivation)
 
# 3. Wellness score (higher = healthier)
df['Wellness_Score'] = (
    df['Sleep_Quality_Last_Night_1to5'] +
    df['Mood_Today_1to5'] +
    df['Overall_Energy_Level'] +
    df['Hydration_Glasses_Today'].clip(0, 8) / 2 +
    df['Exercise_Frequency']
    - df['Stress_Level_1to5']
    - df['Junk_Food_Frequency_1to5'] / 2
)
 
# 4. PCOS risk flag (combines known signals)
df['PCOS_Risk_Flag'] = (
    (df['PCOS_PCOD_Diagnosed'] == 1) |
    (
        (df['BMI'] > 27) &
        (df['Periods_Regular'] == 0) &
        (df['Skin_Condition_During_Cycle'] >= 2)
    )
).astype(int)
 
# 5. Period delayed flag (stress-linked cycle prediction feature)
df['Period_Delayed'] = (df['Days_Since_Last_Period'] > 35).astype(int)
 
print("Engineered features added:")
new_feats = ['Cycle_Irregularity_Score','Hormonal_Stress_Index',
             'Wellness_Score','PCOS_Risk_Flag','Period_Delayed']
print(df[new_feats].describe().round(2))
 
```

```python
feature_cols = [c for c in df.columns if c not in [
    'PCOS_Risk_Flag', 'Period_Delayed', 'Is_Anomaly',
    'PCOS_PCOD_Diagnosed',  # direct leak for PCOS target
]]
 
print("=== Top 15 Features for PCOS Risk Prediction ===")
X_pcos = df[feature_cols]
y_pcos = df['PCOS_Risk_Flag']
selector_pcos = SelectKBest(f_classif, k=15)
selector_pcos.fit(X_pcos, y_pcos)
top_pcos = pd.Series(selector_pcos.scores_, index=feature_cols)\
             .sort_values(ascending=False).head(15)
print(top_pcos.to_string())
 
print("\n=== Top 15 Features for Period Delay Prediction ===")
X_delay = df[feature_cols]
y_delay = df['Period_Delayed']
selector_delay = SelectKBest(f_classif, k=15)
selector_delay.fit(X_delay, y_delay)
top_delay = pd.Series(selector_delay.scores_, index=feature_cols)\
              .sort_values(ascending=False).head(15)
print(top_delay.to_string())
```

```python
top_features = list(set(top_pcos.index.tolist() + top_delay.index.tolist()))
 
# Always include engineered features
always_include = [
    'Age', 'BMI', 'Stress_Level_1to5', 'Sleep_Duration',
    'Life_Phase', 'Current_Cycle_Phase', 'Days_Since_Last_Period',
    'Cycle_Irregularity_Score', 'Hormonal_Stress_Index', 'Wellness_Score',
    'PCOS_Risk_Flag', 'Period_Delayed', 'Is_Anomaly'
]
 
final_cols = list(set(top_features + always_include))
df_final = df[final_cols].copy()
 
print("Final ML-ready dataframe shape:", df_final.shape)
print("Columns:", sorted(df_final.columns.tolist()))
```

```python
scale_exclude = ['PCOS_Risk_Flag', 'Period_Delayed', 'Is_Anomaly',
                 'Life_Phase', 'Current_Cycle_Phase']
 
scale_cols = [c for c in df_final.columns if c not in scale_exclude]
 
scaler = StandardScaler()
df_scaled = df_final.copy()
df_scaled[scale_cols] = scaler.fit_transform(df_final[scale_cols])
 
print("Scaling done ✅")
print("Sample scaled values:")
df_scaled[scale_cols[:5]].describe().round(3)
```

```python
plt.figure(figsize=(18, 14))
corr = df_final.drop(columns=['Is_Anomaly']).corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=False, cmap='RdPu',
            linewidths=0.3, vmin=-1, vmax=1)
plt.title('HerLife Feature Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('herlife_correlation_heatmap.png', dpi=150)
plt.show()
print("Heatmap saved ✅")
 
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
 
colors = ['#e91e8c', '#f7a8d4']
 
df_final['PCOS_Risk_Flag'].value_counts().plot(
    kind='bar', ax=axes[0], color=colors, edgecolor='white')
axes[0].set_title('PCOS Risk Flag Distribution', fontweight='bold')
axes[0].set_xticklabels(['No Risk (0)', 'At Risk (1)'], rotation=0)
axes[0].set_xlabel('')
 
df_final['Period_Delayed'].value_counts().plot(
    kind='bar', ax=axes[1], color=colors, edgecolor='white')
axes[1].set_title('Period Delayed Distribution', fontweight='bold')
axes[1].set_xticklabels(['Not Delayed (0)', 'Delayed (1)'], rotation=0)
axes[1].set_xlabel('')
 
plt.tight_layout()
plt.savefig('herlife_class_distribution.png', dpi=150)
plt.show()
 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
 
# Sklearn
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    accuracy_score, f1_score, roc_auc_score, roc_curve,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
 
# XGBoost
from xgboost import XGBClassifier, plot_importance
 
import joblib  # to save models
 
print("All imports successful ✅")
```

```python
df = pd.read_csv('/kaggle/working/herlife_cleaned_full.csv')
print("Loaded shape:", df.shape)
df.head(3)
```

```python
drop_cols = ['User_Journal_Entry', 'Notification_Preference',
             'Most_Energetic_Phase', 'Menstrual_Product_Used']
df.drop(columns=[c for c in drop_cols if c in df.columns], inplace=True)
 
# Fill nulls
fill_none = ['Flow_Intensity','Period_Symptoms','PMS_Symptoms',
             'Current_Cycle_Phase','Hair_Fall_Level',
             'Skin_Condition_During_Cycle','PCOS_PCOD_Symptoms',
             'Other_Medical_Diagnosis']
for col in fill_none:
    if col in df.columns:
        df[col] = df[col].fillna('None')
 
# Ordinal maps
ordinal_maps = {
    'Exercise_Frequency':        {'Never':0,'Sometimes':1,'Regularly':2},
    'Sleep_Duration':            {'Less than 5 hrs':0,'5-7 hrs':1,'7-9 hrs':2,'More than 9 hrs':3},
    'Flow_Intensity':            {'None':0,'Light':1,'Moderate':2,'Heavy':3,'Very Heavy':4},
    'Overall_Energy_Level':      {'Low':0,'Medium':1,'High':2},
    'Missed_Periods_Frequency':  {'Never':0,'Occasionally':1,'Frequently':2},
    'Caffeine_Intake':           {'Never':0,'Rarely':1,'Sometimes':2,'Daily':3},
    'Hair_Fall_Level':           {'None':0,'Mild':1,'Moderate':2,'Severe':3},
    'Skin_Condition_During_Cycle':{'Clear':0,'Mild breakout':1,'Moderate breakout':2,'Severe breakout':3},
}
for col, mapping in ordinal_maps.items():
    if col in df.columns:
        df[col] = df[col].map(mapping).fillna(0).astype(int)
 
# Label encode nominal columns
le = LabelEncoder()
for col in ['Life_Phase','Diet_Type','Current_Cycle_Phase','Other_Medical_Diagnosis']:
    if col in df.columns:
        df[col] = le.fit_transform(df[col].astype(str))
 
# Binary YES/NO columns
binary_map = {'Yes':1,'No':0,'YES':1,'NO':0,'Not sure':0,'Not applicable':0}
for col in ['Periods_Regular','Clotting','Difficulty_Losing_Weight',
            'Productivity_Changes_With_Cycle','Hormonal_Contraceptive_Use']:
    if col in df.columns:
        df[col] = df[col].map(lambda x: binary_map.get(str(x).strip(), x))
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
 
# Expand symptom columns
def expand_symptoms(df, col, prefix):
    if col not in df.columns: return df
    dummies = df[col].fillna('None').str.get_dummies(sep=';')
    dummies.columns = [f"{prefix}_{c.strip().replace(' ','_')}" for c in dummies.columns]
    if f"{prefix}_None" in dummies.columns:
        dummies.drop(columns=[f"{prefix}_None"], inplace=True)
    df.drop(columns=[col], inplace=True)
    return pd.concat([df, dummies], axis=1)
 
df = expand_symptoms(df, 'Period_Symptoms',    'SYM')
df = expand_symptoms(df, 'PMS_Symptoms',       'PMS')
df = expand_symptoms(df, 'PCOS_PCOD_Symptoms', 'PCOS_S')
 
# Clip outliers
for col in ['BMI','Weight_kg','Water_Intake_Litres','Days_Since_Last_Period','Avg_Cycle_Length_days']:
    if col in df.columns:
        Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
        df[col] = df[col].clip(Q1 - 1.5*(Q3-Q1), Q3 + 1.5*(Q3-Q1))
 
# Feature engineering
df['Cycle_Irregularity_Score'] = (
    (1 - df['Periods_Regular']) +
    df['Missed_Periods_Frequency'] +
    df['Avg_Cycle_Length_days'].apply(lambda x: 1 if x < 21 or x > 35 else 0)
)
df['Hormonal_Stress_Index'] = df['Stress_Level_1to5'] * (
    df['Sleep_Duration'].apply(lambda x: 1 if x == 0 else 1 + (2 - x) * 0.3))
df['Wellness_Score'] = (
    df['Sleep_Quality_Last_Night_1to5'] +
    df['Mood_Today_1to5'] +
    df['Overall_Energy_Level'] +
    df['Hydration_Glasses_Today'].clip(0,8)/2 +
    df['Exercise_Frequency']
    - df['Stress_Level_1to5']
    - df['Junk_Food_Frequency_1to5']/2
)
 
# Final fill
df = df.fillna(df.median(numeric_only=True))
 
print("Clean shape:", df.shape)
print("Null values remaining:", df.isnull().sum().sum())
 
 
```

```python
df_pcos = df[df['PCOS_PCOD_Diagnosed'] != -1].copy()
df_pcos['PCOS_Target'] = (df_pcos['PCOS_PCOD_Diagnosed'] == 1).astype(int)
 
# Features most relevant to PCOS prediction
# (based on medical knowledge + feature selection from preprocessing)
pcos_features = [
    'Age', 'BMI', 'Weight_kg',
    'Exercise_Frequency', 'Sleep_Duration', 'Stress_Level_1to5',
    'Avg_Cycle_Length_days', 'Periods_Regular', 'Bleeding_Duration_days',
    'Flow_Intensity', 'Clotting', 'Pain_Level_1to5',
    'Missed_Periods_Frequency', 'Difficulty_Losing_Weight',
    'Junk_Food_Frequency_1to5', 'Sugar_Intake_Frequency_1to5',
    'Caffeine_Intake', 'Water_Intake_Litres', 'Overall_Energy_Level',
    'Skin_Condition_During_Cycle', 'Hair_Fall_Level',
    'Hormonal_Contraceptive_Use',
    'Cycle_Irregularity_Score', 'Hormonal_Stress_Index', 'Wellness_Score',
    # Symptom flags (if they exist after expansion)
]
 
# Add any expanded symptom columns that exist
sym_cols = [c for c in df_pcos.columns if c.startswith('SYM_') or
            c.startswith('PMS_') or c.startswith('PCOS_S_')]
pcos_features = pcos_features + sym_cols
 
# Keep only columns that actually exist
pcos_features = [c for c in pcos_features if c in df_pcos.columns]
 
X_pcos = df_pcos[pcos_features]
y_pcos = df_pcos['PCOS_Target']
 
print("PCOS Dataset shape:", X_pcos.shape)
print("Class distribution:\n", y_pcos.value_counts())
print(f"\nPositive rate: {y_pcos.mean()*100:.1f}%")
```

```python
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_pcos, y_pcos,
    test_size=0.2,        # 80% train, 20% test
    random_state=42,
    stratify=y_pcos       # keeps class ratio same in both splits
)
 
print(f"Training set : {X_train_p.shape[0]} rows")
print(f"Test set     : {X_test_p.shape[0]} rows")
 
 
```

```python
neg = (y_train_p == 0).sum()
pos = (y_train_p == 1).sum()
scale_weight = neg / pos
print(f"scale_pos_weight = {scale_weight:.2f}  (handles class imbalance)")
 
xgb_model = XGBClassifier(
    n_estimators     = 200,       # number of trees
    max_depth        = 5,         # how deep each tree grows
    learning_rate    = 0.05,      # step size — smaller = more careful
    subsample        = 0.8,       # use 80% of rows per tree (prevents overfitting)
    colsample_bytree = 0.8,       # use 80% of features per tree
    scale_pos_weight = scale_weight,
    use_label_encoder= False,
    eval_metric      = 'logloss',
    random_state     = 42,
    n_jobs           = -1
)
 
xgb_model.fit(
    X_train_p, y_train_p,
    eval_set=[(X_test_p, y_test_p)],
    verbose=50   # print progress every 50 trees
)
 
print("\nXGBoost training complete ✅")
 
```

```python
y_pred_p  = xgb_model.predict(X_test_p)
y_prob_p  = xgb_model.predict_proba(X_test_p)[:, 1]
 
print("=" * 50)
print("MODEL A — PCOS RISK CLASSIFIER RESULTS")
print("=" * 50)
print(f"Accuracy  : {accuracy_score(y_test_p, y_pred_p)*100:.2f}%")
print(f"F1 Score  : {f1_score(y_test_p, y_pred_p):.4f}")
print(f"ROC-AUC   : {roc_auc_score(y_test_p, y_prob_p):.4f}")
print()
print("Classification Report:")
print(classification_report(y_test_p, y_pred_p,
      target_names=['No PCOS (0)', 'PCOS Risk (1)']))

```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
 
# Confusion Matrix
cm = confusion_matrix(y_test_p, y_pred_p)
disp = ConfusionMatrixDisplay(cm, display_labels=['No PCOS', 'PCOS Risk'])
disp.plot(ax=axes[0], colorbar=False, cmap='RdPu')
axes[0].set_title('Model A — Confusion Matrix', fontweight='bold', fontsize=13)
 
# ROC Curve
fpr, tpr, _ = roc_curve(y_test_p, y_prob_p)
axes[1].plot(fpr, tpr, color='#e91e8c', lw=2,
             label=f'ROC AUC = {roc_auc_score(y_test_p, y_prob_p):.3f}')
axes[1].plot([0,1],[0,1],'--', color='gray')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('Model A — ROC Curve', fontweight='bold', fontsize=13)
axes[1].legend()
 
plt.tight_layout()
plt.savefig('model_a_evaluation.png', dpi=150)
plt.show()
```

```python
df_period = df[df['Life_Phase'].isin(
    df['Life_Phase'].unique()  # Life_Phase is label-encoded
)].copy()
 
# Recreate target: days until next period
df_period['Days_Until_Next_Period'] = (
    df_period['Avg_Cycle_Length_days'] - df_period['Days_Since_Last_Period']
).clip(lower=0)  # can't be negative
 
# Core features for cycle prediction (what the HerLife app will collect daily)
cycle_features = [
    'Age',
    'Stress_Level_1to5',           # daily log
    'Sleep_Quality_Last_Night_1to5', # daily log
    'Sleep_Duration',
    'Days_Since_Last_Period',        # calendar input
    'Avg_Cycle_Length_days',         # from onboarding questions
    'Mood_Today_1to5',               # daily log
    'Hydration_Glasses_Today',       # daily log
    'Exercise_Frequency',
    'Water_Intake_Litres',
    'Overall_Energy_Level',
    'Caffeine_Intake',
    'Hormonal_Stress_Index',         # engineered feature
    'Cycle_Irregularity_Score',      # engineered feature
    'Wellness_Score',                # engineered feature
    'Bleeding_Duration_days',
    'Flow_Intensity',
    'Periods_Regular',
    'Missed_Periods_Frequency',
    'Pain_Level_1to5',
    'BMI',
    'PCOS_PCOD_Diagnosed',
]
 
cycle_features = [c for c in cycle_features if c in df_period.columns]
 
X_cycle = df_period[cycle_features]
y_cycle = df_period['Days_Until_Next_Period']
 
print("Cycle Predictor Dataset shape:", X_cycle.shape)
print("\nTarget (Days Until Next Period) stats:")
print(y_cycle.describe().round(2))
 
 
```

```python
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_cycle, y_cycle,
    test_size=0.2,
    random_state=42
)
 
# SVR is sensitive to feature scale — StandardScaler is mandatory
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_c)
X_test_scaled  = scaler.transform(X_test_c)   # use same scaler, don't refit!
 
print(f"Train: {X_train_c.shape[0]} rows | Test: {X_test_c.shape[0]} rows")
print("Scaling done ✅")
 
```

```python
svr_model = SVR(
    kernel  = 'rbf',    # Radial Basis Function — best for non-linear health data
    C       = 100,      # regularisation — higher = fits data more closely
    epsilon = 0.5,      # tolerance margin — predictions within ±0.5 days = no penalty
    gamma   = 'scale'   # auto-scales based on number of features
)
 
svr_model.fit(X_train_scaled, y_train_c)
print("SVR training complete ✅")
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train_c)
print("Linear Regression training complete ✅")
```

```python
def evaluate_regression(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2   = r2_score(y_test, y_pred)
    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(f"  MAE  (avg days off)     : {mae:.2f} days")
    print(f"  RMSE (penalises big err): {rmse:.2f} days")
    print(f"  R²   (variance explained): {r2:.4f}  (1.0 = perfect)")
    return y_pred
 
y_pred_svr = evaluate_regression("SVR — Period Predictor",
                                  svr_model, X_test_scaled, y_test_c)
y_pred_lr  = evaluate_regression("Linear Regression — Period Predictor",
                                  lr_model,  X_test_scaled, y_test_c)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
 
for ax, y_pred, title, color in zip(
    axes,
    [y_pred_svr, y_pred_lr],
    ['SVR Predictions', 'Linear Regression Predictions'],
    ['#e91e8c', '#9b59b6']
):
    ax.scatter(y_test_c, y_pred, alpha=0.5, color=color, edgecolors='white', s=40)
    # Perfect prediction line
    mn = min(y_test_c.min(), y_pred.min())
    mx = max(y_test_c.max(), y_pred.max())
    ax.plot([mn, mx], [mn, mx], 'k--', lw=1.5, label='Perfect prediction')
    ax.set_xlabel('Actual Days Until Next Period')
    ax.set_ylabel('Predicted Days Until Next Period')
    ax.set_title(f'Model B — {title}', fontweight='bold')
    ax.legend()
 
plt.tight_layout()
plt.savefig('model_b_predictions.png', dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
 
for ax, y_pred, title, color in zip(
    axes,
    [y_pred_svr, y_pred_lr],
    ['SVR Residuals', 'Linear Regression Residuals'],
    ['#e91e8c', '#9b59b6']
):
    residuals = y_test_c - y_pred
    ax.hist(residuals, bins=30, color=color, edgecolor='white', alpha=0.85)
    ax.axvline(0, color='black', linestyle='--', lw=1.5)
    ax.set_xlabel('Residual (Actual − Predicted) in Days')
    ax.set_ylabel('Count')
    ax.set_title(f'Model B — {title}', fontweight='bold')
 
plt.tight_layout()
plt.savefig('model_b_residuals.png', dpi=150)
plt.show()
```

```python
coef_df = pd.DataFrame({
    'Feature'    : cycle_features,
    'Coefficient': lr_model.coef_
}).sort_values('Coefficient', key=abs, ascending=False)
 
plt.figure(figsize=(10, 7))
colors = ['#e91e8c' if c > 0 else '#9b59b6' for c in coef_df['Coefficient']]
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color=colors)
plt.axvline(0, color='black', lw=0.8)
plt.title('Model B — Feature Coefficients (Linear Regression)',
          fontweight='bold', fontsize=13)
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.savefig('model_b_coefficients.png', dpi=150)
plt.show()
 
print(coef_df.to_string(index=False))
 
```
